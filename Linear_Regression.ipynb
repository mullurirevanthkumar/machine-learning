# **Lineare Regression**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
IceCream = pd.read_csv('/content/sample_data/IceCreamData.csv')
print(IceCream.head(10))
print(IceCream.describe())
print(IceCream.info())
```

# **Visualize Data**

```python
sns.jointplot(x='Temperature', y='Revenue', data=IceCream, color='gray')
```

```python
sns.pairplot(IceCream)
```

```python
sns.lmplot(x = 'Temperature', y = 'Revenue', data = IceCream)
```

```python
X = IceCream[['Temperature']]
y = IceCream[['Revenue']]
```

# **Split The Data**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

print(X_train)
print(y_train)
print(X_train.shape)
print(y_train.shape)
```

# **Train The Model**

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression(fit_intercept= True)
```

```python
regressor.fit(X_train, y_train)

print('Linear model Coefficiant m', regressor.coef_)
print('Linear  B', regressor.intercept_)
```

# **Test The Model**

```python
y_predict = regressor.predict(X_test)
print(y_predict)
print(y_test)
```

```python
plt.scatter(X_train, y_train, color='gray')
plt.plot(X_train, regressor.predict(X_train), color = 'red')
plt.ylabel('Revenue [Dollors]')
plt.xlabel('Temperature [DegC]')
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, regressor.predict(X_test), color = 'red')
plt.ylabel('Revenue [Dollors]')
plt.xlabel('Temperature [DegC]')
```

## **Fuel Consumption**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fuel_economy = pd.read_csv('/content/sample_data/FuelEconomy.csv')
fuel_economy.head(5)
```

```python
print(fuel_economy.describe())
print(fuel_economy.info())
```

```python
sns.jointplot(x='Horse Power', y='Fuel Economy (MPG)', data = fuel_economy)
```

```python
from sklearn.model_selection import train_test_split

X= fuel_economy[['Horse Power']]
y= fuel_economy[['Fuel Economy (MPG)']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

print(X_train)
print(y_train)
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression(fit_intercept= True)

regressor.fit(X_train, y_train)
print('Linear model Coefficiant m', regressor.coef_)
print('Linear  B', regressor.intercept_)
```

```python
plt.scatter(X_train, y_train, color='gray')
plt.plot(X_train, regressor.predict(X_train), color = 'red')
plt.ylabel('Horse Power')
plt.xlabel('Fuel Economy (MPG)')
```

```python
plt.scatter(X_test, y_test, color='gray')
plt.plot(X_test, regressor.predict(X_test), color = 'red')
plt.ylabel('Horse Power')
plt.xlabel('Fuel Economy (MPG)')
```
